```python
# We only have one categorical nominal column ("Gender") and two ordinal ("Vehicle Class", "Vehicle Size").
# Therefore, we will need to separate them in order to apply the OneHotEncoder to the categorical nominal variable
# and then we will need to use replace() to cast categorical ordinal variables.

# These are the dataframes that we will have to split in order to do the transformations

#display(X_cat_train_df)
display(X_cat_test_df)

# Then we will have something similar to this...

# Cat train NOMINAL gender df
# Cat train ORDINAL vehicle class and vehicle size df

# Cat test NOMINAL gender df
# Cat test ORDINAL vehicle class and vehicle size df
```

```python
# Create copies of test and train dataframes showing only the categorical
# nominal column ("Gender")

X_cat_nominal_train_df = X_cat_train_df[['Gender']].copy()
X_cat_nominal_test_df = X_cat_test_df[['Gender']].copy()
```

```python
#display(X_cat_nominal_train_df)
display(X_cat_nominal_test_df)
```

```python
# Use OneHotEncoder to transform categorical nominal variables

from sklearn.preprocessing import OneHotEncoder

encoder_cat_nominal = OneHotEncoder(sparse=False)

encoder_cat_nominal.fit(X_cat_nominal_train_df) 

path = "encoders/"
isExist = os.path.exists(path)
if not isExist:
    os.makedirs(path)
    print("The new directory is created!")

encoder_cat_nominal_file_name = "encoder_cat_nominal.pkl" 
with open(path+encoder_cat_nominal_file_name, "wb") as file:
    pickle.dump(encoder_cat_nominal, file) 

X_cat_nominal_train_scaled = encoder_cat_nominal.transform(X_cat_nominal_train_df) # Apply transformation on train set
X_cat_nominal_test_scaled = encoder_cat_nominal.transform(X_cat_nominal_test_df) # Apply transformation on test set

column_cat_nominal = encoder_cat_nominal.get_feature_names_out(['Gender'])
# We had to add the get_feature_names_out() in this case because the transformation returned values for the 
# feminine and masculine columns separately and it was retrieving an error when creating the dataframes (which
# had only one column for gender)
```

```python
# Create dataframes as the transformation only creates numpy arrays - therefore, we will need to add columns
# as well

X_cat_nominal_train_scaled_df = pd.DataFrame(X_cat_nominal_train_scaled, columns=column_cat_nominal)
X_cat_nominal_test_scaled_df = pd.DataFrame(X_cat_nominal_test_scaled, columns=column_cat_nominal)
```

```python
# Display dataframe

#display(X_cat_nominal_train_scaled_df)
display(X_cat_nominal_test_scaled_df)
```

```python
# Create function to compute metrics

def error_metrics_report(y_real_train: list, y_real_test: list, y_pred_train: list, y_pred_test: list) -> pd.DataFrame:
    '''
    This function takes a list of values and returns a Pandas DataFrame collecting all the errors metrics
    
    Inputs:
    
    List of values 
    
    Outputs:
    
    '''
    

    MAE_train = mean_absolute_error(y_real_train, y_pred_train)
    MAE_test  = mean_absolute_error(y_real_test, y_pred_test)

    # Mean squared error
    MSE_train = mean_squared_error(y_real_train, y_pred_train)
    MSE_test  = mean_squared_error(y_real_test, y_pred_test)

    # Root mean squared error
    RMSE_train = mean_squared_error(y_real_train, y_pred_train, squared=False)
    RMSE_test  = mean_squared_error(y_real_test,  y_pred_test,  squared=False)

    # R2
    R2_train = r2_score(y_real_train, y_pred_train)
    R2_test  = r2_score(y_real_test,  y_pred_test)

    results = {"Metric": ["MAE","MSE", "RMSE", "R2"],
               "Train": [MAE_train, MSE_train, RMSE_train, R2_train],
               "Test":  [MAE_test, MSE_test, RMSE_test, R2_test]}

    results_df = pd.DataFrame(results).round(2)

    return results_df
```
